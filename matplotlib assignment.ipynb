```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
import pandas as pd

%matplotlib inline
```

## Question 1

```python
max = [39, 41, 43, 47, 49, 51, 45, 38, 37, 29, 27, 25]
min = [21, 23, 27, 28, 32, 35, 31, 28, 21, 19, 17, 18]
```

```python
temp_fun = lambda time, temp, consts: temp + consts * np.cos((time) * 2 * np.pi / time.max())

popt_max, pcov_max = curve_fit(temp_fun, np.arange(1, 13, 1), max)
popt_min, pcov_min = curve_fit(temp_fun, np.arange(1, 13, 1), min)
```

```python
fig, ax = plt.subplots(1,1)
ax.set_xlabel("Month")
ax.set_ylabel("Temperature(Â°C)")

ax.scatter(range(1,13), max, color="r")

ax.plot(np.arange(1, 13, 1), temp_fun(np.arange(1, 13, 1), *popt_max), 'r')
ax.plot(np.arange(1, 13, 1), temp_fun(np.arange(1, 13, 1), *popt_min), 'b')

ax.scatter(range(1,13), min, color="b")
```
